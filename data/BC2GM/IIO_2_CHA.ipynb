```python
import json
```

```python
with open('train_IIO.json', 'r', encoding='utf-8') as f:
    train_IIO = json.load(f)
with open('dev_IIO.json', 'r', encoding='utf-8') as f:
    dev_IIO = json.load(f)
with open('test_IIO.json', 'r', encoding='utf-8') as f:
    test_IIO = json.load(f)
```

```python
for data in train_IIO:
    data['instruction'] = "Please find all the entity words associated with the GENE category in the given text. Output format is 'Entity_1$Type_2##Entity_2$Type_2...'."+"\ntext:"
for data in dev_IIO:
    data['instruction'] = "Please find all the entity words associated with the GENE category in the given text. Output format is 'Entity_1$Type_2##Entity_2$Type_2...'."+"\ntext:"
for data in test_IIO:
    data['instruction'] = "Please find all the entity words associated with the GENE category in the given text. Output format is 'Entity_1$Type_2##Entity_2$Type_2...'."+"\ntext:"
```

```python
with open('train.json', 'w', encoding='utf-8') as f:
    json.dump(train_IIO, f, ensure_ascii=False, indent=4)
with open('dev.json', 'w', encoding='utf-8') as f:
    json.dump(dev_IIO, f, ensure_ascii=False, indent=4)
with open('test.json', 'w', encoding='utf-8') as f:
    json.dump(test_IIO, f, ensure_ascii=False, indent=4)
```

```python
def trans_data(source_data):
    target_data = []
    for i, tI in enumerate(source_data):
        train_data_item = {
            "conversation_id": i,
            "category": "BC2GM",
            "conversation": []
        }
        conversation_item = {
            "human": tI['instruction'] + '\nsentence: ' + tI['input'],
            "assistant": tI['output'],
        }
        train_data_item["conversation"].append(conversation_item)
        target_data.append(train_data_item)
    return target_data
    
```

```python
train_data = trans_data(train_IIO)
dev_data = trans_data(dev_IIO)
test_data = trans_data(test_IIO)
```

```python
with open('train_data.json', 'w', encoding='utf-8') as f:
    json.dump(train_data, f, ensure_ascii=False, indent=4)
with open('dev_data.json', 'w', encoding='utf-8') as f:
    json.dump(dev_data, f, ensure_ascii=False, indent=4)
with open('test_data.json', 'w', encoding='utf-8') as f:
    json.dump(test_data, f, ensure_ascii=False, indent=4)
```
